# Necessary Imports & Extracting Data

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from tabulate import tabulate
from bot.explore import get_contents, generate_counts
```

```python
%matplotlib inline
contents = get_contents("data/qa.csv") # Python List
data = generate_counts(contents)
print(data)
```

# Graphical analysis

### Counts per file ### 

```python
plt.xticks(rotation=90)
s = sns.barplot(x="names", y="count",data=data)
```

* This bar chart shows which files (out of the ones which had hits) were the most popular. 

* As one can see, there were a lot of questions asking for permission, as well as unknown questions. 

### Count densities ### 

```python
sns.distplot(data["count"])
```

* The above shows the majority of questions about a particular page was approximately in the 1-8 range. 

```python
sns.kdeplot(data['count'], shade=True)
```
